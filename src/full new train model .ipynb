```python
import tensorflow as tf
from tensorflow import keras
from keras import layers

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import os
from tqdm.notebook import tqdm

from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python

# Load the data
url = "https://raw.githubusercontent.com/Rheyhan/Proyek-TSA/main/Data/Encoded/upsampled.csv"
df = pd.read_csv(url)
```

```python

# Define features and target variable -> disini dicoba gapake kelurahan
X = df[['RAIN_D-1', 'RAINFALL_D-1', 'RAIN_LAST3D', 'ARAINFALL_LAST3D', 'RAIN_LAST7D', 'RAINFALL_LAST7D', 'TEPS_OF_DAY', 'CIPINANG HULU', 'SUNTER HULU', 'PULO GADUNG']]
y = df['Kejadian']

```

```python
# Convert target into LabelEncoder
encoder = LabelEncoder()
y1 = encoder.fit_transform(y)

# Convert target into one hot encoding
Y = to_categorical(y1)

# Convert X and Y into train and test data
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
# Define a model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(2, activation='softmax')
])

# Compile the model
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, batch_size=50, epochs=100)
```

```python
# Evaluate the model with test data
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print('Test loss:', loss)
print('Test accuracy:', accuracy)

```

```python
# Predict test data
y_pred = model.predict(X_test)

# Print actual and predicted value
actual = np.argmax(y_test, axis=1)
predicted = np.argmax(y_pred, axis=1)
print(f"Actual: {actual}")
print(f"Predicted: {predicted}")
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

# Calculate the confusion matrix
confusion = confusion_matrix(actual, predicted)
print("Confusion Matrix:")
print(confusion)

# Calculate accuracy
accuracy = accuracy_score(actual, predicted)
print("Accuracy:", accuracy)

# Calculate precision
precision = precision_score(actual, predicted)
print("Precision:", precision)

# Calculate recall
recall = recall_score(actual, predicted)
print("Recall:", recall)

# Calculate F1-score
f1 = f1_score(actual, predicted)
print("F1-Score:", f1)
```
